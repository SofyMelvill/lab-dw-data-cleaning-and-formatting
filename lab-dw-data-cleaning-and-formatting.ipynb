## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
insur_comp_df.head()
```

```python
# I will start with the state variable

insur_comp_df["state"] = insur_comp_df["state"].str.strip()
insur_comp_df["state"] = insur_comp_df["state"].replace({
    "Cali" : "California",
    "CALI" : "California",
    "WA" : "Whashington",
    "Washington" : "Whashington",
    "AZ" : "Arizona"
})

print(insur_comp_df.value_counts(["state"]))
print(insur_comp_df["state"].unique())
```

```python
print(insur_comp_df.value_counts(["gender"]))
print(insur_comp_df["gender"].unique())
```

```python
# Clean spaces, lower letters and make it standard

insur_comp_df["gender"] = insur_comp_df["gender"].str.strip()
insur_comp_df["gender"] = insur_comp_df["gender"].replace({
    "Male" : "M",
    "female" : "F",
    "Femal" : "F"
})
insur_comp_df.loc[~insur_comp_df["gender"].isin(["M", "F"]), "gender"] = pd.NA

print(insur_comp_df.value_counts(["gender"]))
print(insur_comp_df["gender"].unique())
```

```python
insur_comp_df["education"] = insur_comp_df["education"].str.strip().str.title()

insur_comp_df["education"] = insur_comp_df["education"].replace("Bachelors", "Bachelor")

print(insur_comp_df.value_counts(["education"]))
print(insur_comp_df["education"].unique())
```

```python
# remove % 
insur_comp_df["customer_lifetime_value"] = insur_comp_df["customer_lifetime_value"].replace('%', '')


insur_comp_df["customer_lifetime_value"]
```

```python
# *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

print(insur_comp_df.value_counts(["vehicle_class"]))
print(insur_comp_df["vehicle_class"].unique())
```

```python
insur_comp_df["vehicle_class"] = insur_comp_df["vehicle_class"].str.strip()
insur_comp_df["vehicle_class"] = insur_comp_df["vehicle_class"].replace({
    "Luxury SUV" : "Luxury",
    "Luxury Car" : "Luxury",
    "Sports Car" : "Luxury"
})

print(insur_comp_df.value_counts(["vehicle_class"]))
print(insur_comp_df["vehicle_class"].unique())
```

```python
insur_comp_df.head()
```

```python
insur_comp_df.describe()
```

```python
insur_comp_df.describe(include=object)
```

```python
insur_comp_df["customer_lifetime_value"] = insur_comp_df["customer_lifetime_value"].astype(str).str.rstrip('%').astype(float)

insur_comp_df["customer_lifetime_value"]
```
